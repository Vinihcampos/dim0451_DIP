```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/laranjas.jpg', 0)

plt.figure(figsize=(15,10))
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.show()
```

```python
_, thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(15,10))
plt.imshow(thresh,cmap='gray')
plt.axis('off')
plt.show()
```

```python
kernel = np.ones((7,7),np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(15,10))
plt.imshow(closing,cmap='gray')
plt.axis('off')
plt.show()
```

```python
kernel = np.ones((5,5),np.uint8)
erode = cv2.erode(closing,kernel,iterations=1)
kernel = np.ones((7,7),np.uint8)
closing = cv2.morphologyEx(erode, cv2.MORPH_CLOSE, kernel)

#kernel = np.ones((3,3),np.uint8)
#dilation = cv2.dilate(closing,kernel,iterations=5)
#kernel = np.ones((5,5),np.uint8)
#erode = cv2.erode(dilation,kernel,iterations=5)

plt.figure(figsize=(15,10))
plt.imshow(closing,cmap='gray')
plt.axis('off')
plt.show()
```
